```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("D:\Computer-Vision\Image-Processing\\noise\\noise\images.jpg"),cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.hist(cv2.cvtColor(img,cv2.COLOR_RGB2GRAY).reshape(-1),bins=100)
plt.show()
```

Making Gaussian Noise.

```python
mean = 0
std_dev = 10
gaussian_noise = np.random.normal(mean, std_dev, img.shape)
```

```python
plt.hist(gaussian_noise.reshape(-1),bins=100)
plt.show()
```

```python
gaussian_noise.dtype
```

```python
noisy_img = img + gaussian_noise
noisy_img.dtype
```

```python
noisy_img.max()
```

```python
noisy_img.min()
```

Changing the type of image from float to int with Clip function.

```python
noisy_img = np.clip(noisy_img, 0, 255).astype(np.uint8)
```

```python
plt.hist(cv2.cvtColor(noisy_img,cv2.COLOR_RGB2GRAY).reshape(-1),bins=100)
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.title("Noisy Image")
plt.imshow(noisy_img)
plt.show()
```

Denoising by Average filter (Blur filter) wirh filter (3,3).

One of the ways to denoise the Noise made by Gaussian Noise, is Using Blur filter.

```python
img_denoise = cv2.blur(noisy_img,(3,3))
```

```python
plt.title("Denoised by Blur filter Image")
plt.imshow(img_denoise)
plt.show()
plt.title("Noisy Image")
plt.imshow(noisy_img)
plt.show()
```

Now trying denoise progress with Gaussian blur filter. There is an additional option named Sigma and as much as we increase it we will lost ,ore information and we does more deniose. 

```python
img_denoise = cv2.GaussianBlur(noisy_img,(11,11),0.6)
```

```python
plt.imshow(img_denoise)
plt.show()
```

Denoising with median Blur filter.

Its the best choise for the salt and pepper noise.

3 >>> filter size.

```python
img_denoise = cv2.medianBlur(noisy_img,3)
```

```python
plt.imshow(img_denoise)
plt.show()
```

BilateralFilter is a suitable choise for gaussian niose image, but is not a good choise for slat-pepper images.

```python
img_denoise = cv2.bilateralFilter(noisy_img,11,sigmaColor=20,sigmaSpace=10)
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.title("noisy Image")
plt.imshow(noisy_img)

plt.subplot(1, 2, 2)
plt.title("Image Denoise")
plt.imshow(img_denoise)
plt.show()
```
