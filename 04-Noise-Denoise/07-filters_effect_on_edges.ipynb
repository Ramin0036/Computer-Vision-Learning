```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("D:\Computer-Vision\Image-Processing\\noise\\noise\shape.jpg"),cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

Crossing from black pixel with Intensity = 0 to Withe pixel with Intensity = 255. Its a clear edge.

```python
plt.imshow(img[:100,:100,:])
plt.show()
```

Filter Effect.

```python
img_blur = cv2.blur(img,(7,7))
```

Bluring the image with blur filter.

```python
plt.imshow(img_blur)
plt.show()
```

If you see we dosnt have clear edge. Filters are going to smoothing the edges.

```python
plt.imshow(img_blur[:100,250:350,:])
plt.show()
```

```python
img_gaussian = cv2.GaussianBlur(img,(11,11),2)
```

```python
plt.imshow(img_gaussian)
plt.show()
```

And there is a other example with bigger filter. If we can not recogunizing the edges, the model also can not recogunie.

```python
plt.imshow(img_gaussian[40:60,40:60])
plt.show()
```

```python
img_median = cv2.medianBlur(img,3)
```

If you turn back and see the org image you can see a line in down right corner of image but median filter complitely wiped out this one.

```python
plt.imshow(img_median)
plt.show()
```

```python
plt.imshow(img_median[40:60,40:60])
plt.show()
```

```python
plt.imshow(img_median[20:100,270:330,:])
plt.show()
```

Minimizing the filter harms to edges.

```python
img_bilateral = cv2.bilateralFilter(img,11,sigmaColor=50,sigmaSpace=10)
```

```python
plt.imshow(img_bilateral)
plt.show()
```

```python
plt.imshow(img_bilateral[40:60,40:60])
plt.show()
```
