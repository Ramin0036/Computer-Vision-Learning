```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.cvtColor((cv2.imread("D:\Computer-Vision\Image-Processing\data\\blue.jpg")) , cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(image)
plt.show()
```

```python
R,G,B = cv2.split(image)
```

```python
# Gamma correction for Blue channel
```

```python
B.max()
```

```python
B.min()
```

```python
plt.hist(B.reshape(-1) , bins=100)
plt.show()
plt.hist(G.reshape(-1) , bins=100)
plt.show()
plt.hist(R.reshape(-1) , bins=100)
plt.show()
```

```python
gamma = 8
B_new = ((np.power((B / 255) , gamma) * 255)).astype('uint8')
```

```python
plt.hist(B.reshape(-1) , bins=100)
plt.show()
plt.hist(B_new.reshape(-1) , bins=100)
plt.show()
```

```python
image_new = cv2.merge([R,G,B_new])
```

```python
plt.imshow(image)
plt.show()
plt.imshow(image_new)
plt.show()
```

```python
#Changing saturation.
```

```python
image_HSV = cv2.cvtColor(image , cv2.COLOR_RGB2HSV)
```

```python
H,S,V = cv2.split(image_HSV)
```

```python
plt.hist(S.reshape(-1) , bins=100)
plt.show()
```

```python
gamma = 0.1
S_new = ((np.power((S / 255) , gamma) * 255)).astype('uint8')
image_HSV_new = cv2.merge([H,S_new,V])
```

```python
image_HSV_Showable = cv2.cvtColor(image_HSV_new,cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(image_HSV_Showable)
plt.show()
```
