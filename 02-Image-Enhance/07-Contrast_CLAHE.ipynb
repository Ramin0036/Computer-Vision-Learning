```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("D:\Computer-Vision\Image-Processing\data\lowcontrast3.jpg"),cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(img)
plt.show()
```

```python
h,s,v = cv2.split(img_hsv)
```

```python
v.min()
```

```python
v.max()
```

Using CLAHE. clipLimit=Contrast limit (CL part) to 2 frequancy. tileGridSize is AHE part.

Higher clipLimit >>> Higher Contrast.

```python
clahe = cv2.createCLAHE(clipLimit=2,tileGridSize=(2,2))
v_clahe = clahe.apply(v)
```

```python
v_clahe.min()
```

```python
v_clahe.max()
```

```python
plt.hist(v.reshape(-1),bins=100)
plt.xlim(0,255)
plt.ylim(0,30000)
plt.show()
plt.hist(v_clahe.reshape(-1),bins=100)
plt.xlim(0,255)
plt.ylim(0,30000)
plt.show()
```

```python
img_clahe = cv2.merge([h,s,v_clahe])
```

```python
def img_hsv_show(img_hsv):
    img_show = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
    plt.imshow(img_show)
    print(f'image shape: {img_show.shape}')
    plt.show()
```

```python
plt.imshow(img)
plt.show()
img_hsv_show(img_clahe)
```

```python
# color image - RGB
chlahe_channels = []
img_channel = cv2.split(img)
for ch in img_channel:
    clahe = cv2.createCLAHE(clipLimit=2,tileGridSize=(2,2))
    clahe_ch = clahe.apply(ch)
    chlahe_channels.append(clahe_ch)

img_rgb_clahe = cv2.merge(chlahe_channels)
```

```python
plt.imshow(img)
plt.show()
plt.imshow(img_rgb_clahe)
plt.show()
```
