```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("D:\Computer-Vision\Image\data\image\lowcontrast2.jpg") , cv2.COLOR_BGR2RGB)
```

HSV image.

```python
img_HSV = cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(img)
plt.show()
```

Stretching on HSV.

```python
h,s,v = cv2.split(img_HSV)
```

```python
plt.hist(v.reshape(-1),bins=100)
plt.show()
```

```python

v_min = v.min()
v_max = v.max()
v_stretched = ((v - v_min) / (v_max - v_min) * 255).astype('uint8')
```

```python
HSV_img_stretched = cv2.merge([h,s,v_stretched])
```

```python
HSV_img_stretched = cv2.cvtColor(HSV_img_stretched , cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(HSV_img_stretched)
plt.show()
plt.imshow(img)
plt.show()
```

Stretched image is darker than the original image. Its because of that we probabily have a point in high value like 220 limmiting our stretch. Then its going to stretch to lower values.

```python
v_equ = cv2.equalizeHist(v)
```

```python
plt.hist(v_equ.reshape(-1),bins=100)
plt.show()
```

```python
HSV_img_equ = cv2.merge([h,s,v_equ])
```

```python
HSV_img_equ = cv2.cvtColor(HSV_img_equ , cv2.COLOR_HSV2RGB)
```

```python
plt.imshow(HSV_img_equ)
plt.show()
plt.imshow(HSV_img_stretched)
plt.show()
plt.imshow(img)
plt.show()
```
