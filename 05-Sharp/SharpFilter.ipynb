```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("D:\Computer-Vision\Image-Processing\Edges\sharp\old.jpeg"),cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(15,6))
plt.imshow(img)
plt.show()
```

Filters for sharping the image. For getting more weight in center pixel giving a bigger number than the others for lossing the weight for the other pixels.

```python
sharpen_filter1 = np.array([[ 0, -1,  0],
                            [-1,  5, -1],
                            [ 0, -1,  0]])

sharpen_filter2 = np.array([[ -1, -1,  -1],
                            [ -1,  9,  -1],
                            [ -1, -1,  -1]])
```

```python
img_sharp1 = cv2.filter2D(img,-1,sharpen_filter1)
```

```python
plt.figure(figsize=(15,6))
plt.imshow(img)
plt.figure(figsize=(15,6))
plt.imshow(img_sharp1)
plt.show()
```

```python
img_sharp2 = cv2.filter2D(img,-1,sharpen_filter2)
```

```python
plt.figure(figsize=(15,6))
plt.imshow(img)
plt.figure(figsize=(15,6))
plt.imshow(img_sharp1)
plt.figure(figsize=(15,6))
plt.imshow(img_sharp2)
plt.show()
```
