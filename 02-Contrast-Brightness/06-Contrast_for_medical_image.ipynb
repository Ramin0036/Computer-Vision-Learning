```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.cvtColor(cv2.imread("D:\\Computer-Vision\\Image-Processing\\data\\xray.jpeg"),cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(img)
plt.show()
```

```python
# Histogram Stretching
```

```python
h,s,v = cv2.split(img_hsv)
```

```python
v_min = v.min()
v_max = v.max()
```

```python
v_min
```

```python
v_max
```

```python
plt.hist(v.reshape(-1),bins=100)
plt.show()
```

```python
v_stretch = (((v - v_min) / (v_max - v_min)) * 255).astype('uint8')
```

```python
v_stretch.max()
```

```python
v_stretch.min()
```

```python
plt.hist(v_stretch.reshape(-1),bins=100)
plt.show()
```

```python
img_hsv_stretched = cv2.merge([h,s,v_stretch])
```

```python
def img_hsv_show(img_hsv):
    img_show = cv2.cvtColor(img_hsv,cv2.COLOR_HSV2RGB)
    plt.imshow(img_show)
    print(f'image shape: {img_show.shape}')
    plt.show()
```

```python
plt.imshow(img)
plt.show()
```

```python
img_hsv_show(img_hsv_stretched)
```

```python
# Histogram Equalization
```

```python
v_equ = cv2.equalizeHist(v)
```

```python
plt.hist(v_equ.reshape(-1),bins=100)
plt.show()
```

```python
img_hsv_equ = cv2.merge([h,s,v_equ])
```

```python
plt.imshow(img)
plt.show()
```

```python
img_hsv_show(img_hsv_equ)
```
